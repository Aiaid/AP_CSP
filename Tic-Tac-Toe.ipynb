```python
import random
import functools
LIMIT=1000

def show(game):
    print("\'O\' is computer")
    print("\'X\' is you")
    for i in range(3):
        for j in range(3):
            if (game[i*3+j]==1):
                print("O",end=' ')
            elif (game[i*3+j]== -1):
                print("X", end=' ')
            else :
                print("-",end=' ')
        print("")
    return

def is_win(game):
    for i in  range(3):
        if game[i*3+0]+game[i*3+1]+game[i*3+2]==3:
            return 1
        if game[i*3+0]+game[i*3+1]+game[i*3+2]==-3:
            return -1
        if game[0+i]+game[3+i]+game[6+i]==3:
            return 1
        if game[0+i]+game[3+i]+game[6+i]==-3:
            return -1
    if game[0]+game[4]+game[8]==3:
        return 1
    if game[0]+game[4]+game[8]==-3:
        return -1
    if game[2]+game[4]+game[6]==3:
        return 1
    if game[2]+game[4]+game[6]==-3:
        return -1
    return 0



def human_move(game):
    #show(game)
    while True:
        x=int(input("input your move(1-9): "))
        if (game[x-1]==0):
            game[x-1]=-1
            return game
        else:
            print("invalid move, try again")

def legal_move(game):
    lm=[]
    for i in range(9):
        if game[i]==0:
            lm.append(i)
    return lm


def random_play(game):
    turn=-1
    while True:
        if is_win(game)!=0:
            #show(game)
            return is_win(game)
        lm=legal_move(game)
        if len(lm)==0:
            #show(game)
            return 0
        r=random.randint(0,8)
        while lm.count(r)==0:
            r=random.randint(0,8)
        game[r]=turn
        turn=-turn
            

def random_playout(game):
    total_lose=0
    total_win=0
    for i in range(LIMIT):
        ngame=game.copy()
        r=random_play(ngame)
        if r==-1:
            total_lose+=1
        elif r==1:
            total_win+=1
    return total_win,total_lose


def make_move(game):
    lm=legal_move(game)
    #print(lm)
    best=lm[0]
    best_win=0
    best_lose=LIMIT
    for i in lm:
        ngame=game.copy()
        ngame[i]=1
        w,l=random_playout(ngame)
        if l==best_lose:
            if w>best_win:
                best=i
                best_lose=l
                best_win=w
        if l<best_lose:
            best_lose=l
            best_win=w
            best=i
        #print(i+1,", w= ",w,", l= ",l)
    print("computer make move on :",best+1)
    game[best]=1
    return game
    
def play_game(x):
    game=[0]*9
    turn=random.randint(0,1)
    if x=="h":
        turn=-1
    else :
        turn=1
    while True:
        show(game)
        if is_win(game)==1:
            print("Computer Win")
            break
        elif is_win(game)==-1:
            print("You Win")
            break
        elif len(legal_move(game))==0:
            print("Tie")
            break
        if turn==1:
            make_move(game)
            turn=-turn
        else :
            human_move(game)
            turn=-turn
        
def play_a_new_game():
    while True:
        x=input("who first human or computer(h/c)?")
        while not(x=="h" or x=="c"):
            x=input("who first human or computer(h/c)?")
        play_game(x)
        x=input("play aonther game(y/n)?")
        if x!="y":
            break
```
